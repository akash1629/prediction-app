```python
df['Size']=df['Size'].str.replace('M','000')
df['Size']=df['Size'].str.replace('k','')
```

```python
#there is "'24000', 'Varies with device', '9.4000', '15000', '10000', varied place word
#to remove this replce it with nan
df["Size"].replace('Varies with device',np.nan)
```

```python
char=['+',",",]
col='Installs','Price'
for c in char:
    for i in col:
        df[i]=df[i].str.replace(c,'')
```
