```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("height-weight.csv")
```

now scatter the plot of csv data using plt.scatter(),then label to each axis,xlabel(weight),ylabel(height)

```python
plt.scatter(df["Weight"],df["Height"])
plt.xlabel("weight")
plt.ylabel("height")
```

Now we have to make best fit line on this dotted graph.

just do some modeliing and we have a best fit line.

```python
#lets assign from graph x is weight a (independent feature),y is height dependent
x=df["Weight"]
y=df["Height"]
```

```python
#now put model and do train-test split
from sklearn.model_selection import train_test_split
#after importing, train test split use variables x_train,ytrain,xtest,ytest into train_test_split function or method
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#check size of x,y using shape
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
#as units are different for height and weight, we need to scale the data using standardization,standard scaler
from sklearn.preprocessing import StandardScaler
StandardScaler()
#now save it in a variable
scaler=StandardScaler()
```

```python
#now do put this standardization on training data (only, no test data scaling)
x_train=scaler.fit_transform(x_train.values.reshape(-1, 1))
#this fit_transform is nothing but fit=getting mean and standard deviation of training data, and 
# transform=transforming the training data using this mean and standard deviation
#combine fit_transform means calculate meand and std.devi and then apply that on data.
#we need to reshape x_train because it is a pandas series and not in dataframe or table form
# so for this we need reshape.
x_train
#the output you will see is nothing but standardize dataset of x_train
```

```python
#but we cannot apply this fit_transform on x_test directly, because data leakage will occur so 
#  fit is not use on x_test, but only transform is used
# so we will use scaler.transform on x_test
scaler.transform(x_test.values.reshape(-1, 1))
```
