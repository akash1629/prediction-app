```python
x=df.drop(columns=['FWI','day', 'month', 'year','DC'])
y=df['FWI']
```

```python
#traintest split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
#now here look carefully for selection of features is important before training the model
#there are many features that are correlated to each other those independent ones,
#our target is y so any depend on y is ok, but if within x features they are not ok to be realted ,as no meaning of them if they are similar
#or related or close to each other
#we have to drop correlateed features 
df.corr()
x_train.corr()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(x_train.corr(),annot=True)
```

```python
#make a matrix of this corr
def correlation(data,threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = data.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:  # threshold for correlation
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## threshold--Domain expertise
corr_features=correlation(x_train,0.85)
```

```python
## drop features when correlation is more than 0.85 
x_train.drop(corr_features,axis=1,inplace=True)
x_test.drop(corr_features,axis=1,inplace=True)
x_train.shape,x_test.shape
```

```python
#do standard scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled=scaler.fit_transform(x_train)
x_tes=scaler.transform(x_test)
```

```python
#train model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
new=model.fit(x_scaled,y_train)
y_pred=new.predict(x_tes)
```

```python
from sklearn.metrics import mean_absolute_error
mea=mean_absolute_error(y_test,y_pred)
mea
```

```python
from sklearn.linear_model import LassoCV
lassocv=LassoCV()
lassocv.fit(x_scaled,y_train)
y_pred=lassocv.predict(x_tes)
```
